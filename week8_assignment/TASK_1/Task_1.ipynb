```python
import pandas as pd
import numpy as np
from datetime import datetime, timedelta

# 1. Load CSV
df = pd.read_csv('/content/yellow_tripdata_2016-01.csv', parse_dates=[
    'tpep_pickup_datetime','tpep_dropoff_datetime'
])

# 2. Clean data (as discussed)
df = df.loc[
    (df.trip_distance.between(0.01, 100)) &
    (df.fare_amount >= 0) &
    (df.total_amount >= 0) &
    (df.passenger_count.between(1,6)) &
    (df.pickup_longitude.between(-75, -72)) &
    (df.pickup_latitude.between(40, 42)) &
    (df.dropoff_longitude.between(-75, -72)) &
    (df.dropoff_latitude.between(40, 42))
].copy()

# Q1. Add â€œRevenueâ€
df['Revenue'] = (
    df.fare_amount + df.extra + df.mta_tax +
    df.improvement_surcharge + df.tip_amount +
    df.tolls_amount + df.total_amount
)

# Q2. Increasing count of total passengers by â€œareaâ€
#    bucket lat/lon to 2 decimal places
df['p_lat_b'] = df.pickup_latitude.round(2)
df['p_lon_b'] = df.pickup_longitude.round(2)
q2 = (
    df.groupby(['p_lat_b','p_lon_b'])
      .passenger_count.sum()
      .reset_index(name='total_passengers')
      .sort_values('total_passengers', ascending=True)
)
print("Q2: Passengers by area (asc):")
print(q2.head(10))

# Q3. Realâ€‘time average fare & average Revenue by Vendor
q3 = (
    df.groupby('VendorID')
      .agg(avg_fare=('fare_amount','mean'),
           avg_revenue=('Revenue','mean'))
      .reset_index()
)
print("\nQ3: Avg fare & avg revenue per Vendor:")
print(q3)

# Q4. Moving count of payments by each payment mode
df = df.sort_values('tpep_pickup_datetime')
df['payment_mode_running_count'] = (
    df.groupby('payment_type')
      .cumcount() + 1
)
print("\nQ4: Sample running counts (payment_type):")
print(df[['payment_type','tpep_pickup_datetime','payment_mode_running_count']].head(10))

# Q5. Highest two gaining vendors on a date with total passengers & distance
date_to_analyze = pd.to_datetime('2016-01-15')
mask = df['tpep_pickup_datetime'].dt.date == date_to_analyze.date()
q5 = (
    df.loc[mask]
      .groupby('VendorID')
      .agg(total_revenue=('Revenue','sum'),
           total_passengers=('passenger_count','sum'),
           total_distance=('trip_distance','sum'))
      .reset_index()
      .sort_values('total_revenue', ascending=False)
      .head(2)
)
print(f"\nQ5: Top 2 vendors on {date_to_analyze.date()}:")
print(q5)

# Q6. Route with the most total passengers
q6 = (
    df.groupby([
        'pickup_longitude','pickup_latitude',
        'dropoff_longitude','dropoff_latitude'
    ])['passenger_count']
      .sum()
      .reset_index(name='passenger_sum')
      .sort_values('passenger_sum', ascending=False)
      .head(1)
)
print("\nQ6: Busiest route by passenger count:")
print(q6)

# Q7. Top pickup locations with most passengers in last N seconds
N = 10  # seconds
cutoff = datetime.utcnow() - timedelta(seconds=N)
# If your tpep times are in UTCâ€”or convert accordinglyâ€”else use pd.Timestamp.now()
recent = df.loc[df.tpep_pickup_datetime >= cutoff]
q7 = (
    recent.groupby(['pickup_longitude','pickup_latitude'])
          .passenger_count.sum()
          .reset_index(name='passenger_sum')
          .sort_values('passenger_sum', ascending=False)
          .head(5)
)
print(f"\nQ7: Top pickup spots in last {N} seconds:")
print(q7)
```
